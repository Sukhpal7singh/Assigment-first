```python
# Q1: making dataframe of job title, location, company_name and experience

# initializing the selenium driver
driver = webdriver.Edge(service=edge_executable)
driver.implicitly_wait(50000)
url = "https://naukri.com/"

# going to the given url
driver.get(url)

# searching for data analyst by location Banglore
inputs = driver.find_elements(by=By.CLASS_NAME, value = 'suggestor-input')
skill = 'Data Analyst'
location = 'Banglore'
skills_input = inputs[0]
location_input = inputs[1]
skills_input.send_keys(skill)
location_input.send_keys(location)
search_button = driver.find_element(by=By.CLASS_NAME, value='qsbSubmit')
search_button.click()
```

```python
# scrap result
# make sure that driver is running in background

job_list = []

# function to scrap the jobs and appending int the list
def scrap_jobs_and_make_web_element():
    job_card_web_elements = driver.find_elements(by=By.CLASS_NAME, value = 'jobTuple')
    for element in job_card_web_elements:
        category = skill
        job_title = element.find_element(by=By.CLASS_NAME, value = 'title').text
        company_name = element.find_element(by=By.CLASS_NAME, value = 'subTitle').text
        job_detail_elements = element.find_elements(by=By.CLASS_NAME, value = 'mt-7')[1]
        job_details = job_detail_elements.find_elements(by=By.TAG_NAME, value = 'span')
        experience = job_details[0].text
        salary = job_details[1].text
        location = job_details[2].text
        job = [job_title, location, company_name, experience, salary]
        job_list.append(job)
    
scrap_jobs_and_make_web_element()
    
```

```python
import csv
from pathlib import Path
```

```python
# some utility functions to reduce repetitive tasks

# function to create the file and do nothing if file already exists
def create_file(file_title):    
    file_path = file_title
    file = Path(file_path)
    file.touch(exist_ok=True)

# this function will delete the given column number from the list
def delete_column_by_number(target_list, col_no):
    for child in target_list:
        del child[col_no]

# this function will create the csv file of the given list
def write_as_csv(title_list, list_to_save,file_name_to_save, save_till_no = 10):
    create_file(file_name_to_save)
    file = open(file_name_to_save, 'w', encoding='utf-8')
    writer = csv.writer(file)
    writer.writerow(title_list)
    i = 0
    for element in list_to_save:
        if i == save_till_no:
            break
        writer.writerow(element)
        i+=1
```

```python
# exporting dataframe as csv file
file_name = 'q1_job_data.csv'
csv_title_list = ['Job_Title', 'location', 'company_name', 'experience']
q1_job_list = deepcopy(job_list)
delete_column_by_number(q1_job_list, 4)
write_as_csv(csv_title_list, q1_job_list, file_name)
```

```python
# this function will generate the required data from given title, description and price list
def generate_product_list(title_element_list, description_element_list, price_element_list):
    title_list_as_text = []
    description_list_as_text = []
    price_list_as_text = []
    
    my_range = (0, len(title_element_list)-1)
    for i in my_range:
        title_element = title_element_list[i]
        title_list_as_text.append(title_element.text)
    for i in my_range:
        description_element = description_element_list[i]
        description_list_as_text.append(description_element.text)
    for i in my_range:
        price_element = price_element_list[i]
        price_list_as_text.append(price_element.text)
    
    title_list_len = len(title_list_as_text)
#     print(title_list_len)
    my_range = (0, title_list_len-1)
    should_append = True
    for i in my_range:
        if should_append == True:
            sunglass = [title_list_as_text[i], description_list_as_text[i], price_list_as_text[i]]
            sunglass_list.append(sunglass)
        should_append = not should_append
```

```python
# elements in which required data is present
elements = driver.find_elements(by=By.CLASS_NAME, value = '_1xHGtK')
len(elements)

# this function will get our data of sunglasses and append it to the sunglass_list
def get_elements():
    for ele in elements:
        title_element_list = ele.find_elements(by=By.CLASS_NAME, value='_2WkVRV')
        description_element_list = ele.find_elements(by=By.CLASS_NAME, value = 'IRpwTa')
        price_element_list = ele.find_elements(by=By.CLASS_NAME, value = '_30jeq3')
        generate_product_list(title_element_list, description_element_list, price_element_list)
```
